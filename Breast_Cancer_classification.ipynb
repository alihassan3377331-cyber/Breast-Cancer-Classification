#Importing libraries

```python
import os
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
path = kagglehub.dataset_download("uciml/breast-cancer-wisconsin-data")
path_breast_cancer = os.path.join(path, "data.csv")
breast_cancer=pd.read_csv(path_breast_cancer)
```

##calling the dataset

```python
breast_cancer
```

#droping the columns

```python
breast_cancer.drop("Unnamed: 32",axis=1,inplace=True)
breast_cancer.drop("id",axis=1,inplace=True)
```

```python
breast_cancer
```

##adding the new columns

```python
data_frame=pd.DataFrame(breast_cancer.values,columns=breast_cancer.columns)
```

```python
data_frame['label']=breast_cancer['diagnosis']
```

```python
breast_cancer
```

##checking the null value

```python
breast_cancer.isnull().sum()
```

#checking the doublicate value

```python
breast_cancer.duplicated()
```

###describtion of the data

```python
breast_cancer.describe()
```

```python
data_frame['label'].value_counts()
```

#convert string to int

```python
breast_cancer.replace({'diagnosis':{'M':1,'B':0}},inplace=True)
```

##spliting the data

```python
y = data_frame['label']
x = data_frame.drop(columns=['diagnosis', 'label'], axis=1)
```

##train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

#model training

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

#prediction

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

#accuracy of logictic regression at y_test





```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

#accuracy of logictic regression at y_train


```python
y_prediction=model.predict(x_train)
```

```python
y_prediction
```

```python
accuracy_score(y_train,y_prediction)
```
